```python
# Use this cell to begin your analysis, and add as many as you would like!
# Import necessarly modules
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use('fivethirtyeight')
```

```python
# Load the dataset
office_episodes = pd.read_csv('datasets/office_episodes.csv')

# To view size of dataset and first five rows
print(office_episodes.shape) # print number of (rows, columns) of our dataset
office_episodes.head()
```

```python
# Lets now prepare our scatter plot
# To prepare the x-axis
episode_number_x = office_episodes['episode_number']

# To prepare the y-axis
viewership_y = office_episodes['viewership_mil']
```

# Question 1

```python
# For Color Schame and Marker Size, we will iterate over every row in our dataset and make colors & sizes lists 
# based on required conditions
colors = []
sizes = [] 

for index, row in office_episodes.iterrows() :
# For colors list    
    if row['scaled_ratings'] < 0.25 :
        colors.append("red")
    elif (row['scaled_ratings'] >= 0.25 and row['scaled_ratings'] < 0.5) :
        colors.append("orange")
    elif (row['scaled_ratings'] >= 0.5 and row['scaled_ratings'] < 0.75) :
        colors.append("lightgreen")
    else:
        colors.append("darkgreen")
   
# For sizes list
    if row['has_guests']:
        sizes.append(250)
    else:
        sizes.append(25)        
```

```python
# Now we are ready for the scatter plot
# Create a figure
fig = plt.figure()
plt.rcParams['figure.figsize'] = [11, 7]
 

# Create a scatter plot
plt.scatter(episode_number_x, viewership_y, s=sizes, c=colors)

#Add Title, x & y labels
plt.xlabel("Episode Number")
plt.ylabel("Viewership (Millions)")
plt.title("Popularity, Quality, and Guest Appearances on the Office")

#Show Plot
plt.show()
```

# Bonus

```python
# Now for the BONUS STEP
# After reading matplotlib.markers documintation, we will add to what we did in Question 1 a markers list, 
# The marker will be circle 'o' for non guest appearances AND star '*' for  guest apperances
markers = []

for index, row in office_episodes.iterrows() :  
    if row['has_guests']:
        markers.append("*")
    else:
        markers.append("o")   
        
# Create a figure
fig = plt.figure()
plt.rcParams['figure.figsize'] = [11, 7]

# Create a scatter plot
for x,y,size,color,mark in zip(episode_number_x,viewership_y,sizes,colors,markers):
    plt.scatter(x, y, s=size, c=color, marker=mark)

#Add Title, x & y labels
plt.xlabel("Episode Number")
plt.ylabel("Viewership (Millions)")
plt.title("Popularity, Quality, and Guest Appearances on the Office")

#Show Plot
plt.show()        
```
