```python
# Use this cell to begin your analyses, and add as many cells as you would like!
# Import pandas
import pandas as pd

# load sets
lego = pd.read_csv('datasets/lego_sets.csv')
parent = pd.read_csv('datasets/parent_themes.csv')
```

```python
# Lets look to lego and see its info
display(lego.info())
display(lego.head())
```

```python
# set_num should not have null values, lets drop all null (set_num) rows
lego_cleaned = lego[~(lego['set_num'].isnull())]
lego_cleaned.info()
```

```python
# Lets now check parent
display(parent.info())
display(parent.head())
```

```python
# Question 1: What percentage of all licensed sets ever released 
# were Star Wars themed?

# First we will filter parent to only licensed once
parent_licensed = parent[parent['is_licensed'] == True]
parent_licensed.head()
```

```python
# Now we will inner join lego_cleaned with parent_licensed 
merged_df = lego_cleaned.merge(parent_licensed, left_on='parent_theme',
                              right_on='name', suffixes=('_lego', '_parent'))
merged_df.head()
```

```python
# merged_df contains the full legos from licensed parents, 
star_is_parent = merged_df['parent_theme'] == 'Star Wars'
star_wars = merged_df[star_is_parent]

# percentage as integer value
the_force = 100 * len(star_wars) // len(merged_df)
the_force
```

```python
# Question 2: In which year was Star Wars not the most popular licensed theme 
# (in terms of number of sets released that year)?

# lets take yearly star wars 
star_wars_yearly = star_wars.groupby('year')[['set_num']].count()
star_wars_yearly.head()
```

```python
# same for non star wars as parent per year
non_star_wars_yearly = merged_df[~star_is_parent].groupby('year'
                                                         )[['set_num']].count()
non_star_wars_yearly.head()
```

```python
# now inner join the two yearly set on 'year'
parent_yearly = star_wars_yearly.merge(non_star_wars_yearly, on='year',
                                      suffixes=('_star', '_not_star'))
parent_yearly.head()
```

```python
# we will add column for the difference between 
# Star Wars and other parents each year
parent_yearly['diff'] = parent_yearly['set_num_not_star'] - parent_yearly['set_num_star']
parent_yearly.head()
```

```python
# The most year Star Wars was not the most pupoluar is the biggest value of 'diff'
# We will sort values descendly
parent_yearly = parent_yearly.sort_values('diff', ascending = False)
parent_yearly.head()
```
